```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#class1
x1 = np.random.normal(25,5,1000)
y1 = np.random.normal(25,5,1000)

#class2
x2 = np.random.normal(10,5,1000)
y2 = np.random.normal(10,5,1000)

#class3
x3 = np.random.normal(55,5,1000)
y3 = np.random.normal(15,5,1000)
```

```python
x = np.concatenate((x1,x2,x3),axis=0)
y = np.concatenate((y1,y2,y3),axis=0)
```

```python
dictionary = {"x":x,"y":y}
```

```python
data = pd.DataFrame(dictionary)
```

```python
data.info()
```

```python
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.scatter(x3,y3)
plt.show()
```

```python
plt.scatter(x1,y1,color = "pink")
plt.scatter(x2,y2,color="purple")
plt.scatter(x3,y3,color ="blue")
plt.show
```

```python
from sklearn.cluster import KMeans
WCSS = []
for k in range(1,15):
      kmeans = KMeans(n_clusters=k)
      kmeans.fit(data)
      WCSS.append(kmeans.inertia_)

plt.plot(range(1,15),WCSS)
plt.xlabel("number of k (cluster) value")
plt.ylabel("wcss")
plt.title("Elbow Method")
plt.grid()
plt.show() 
```

```python
#k=2 iin model 

kmeans2 = KMeans(n_clusters=2)
clusters = kmeans2.fit_predict(data)

data["label"] = clusters 

plt.scatter(data.x[data.label == 0],data.y[data.label == 0],color = "pink")
plt.scatter(data.x[data.label == 1],data.y[data.label == 1],color = "purple")
plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],color = "blue")
plt.show()
```

```python
#k=3 iin model 
kmeans2 = KMeans(n_clusters=3)
clusters = kmeans2.fit_predict(data)

data["label"] = clusters

plt.scatter(data.x[data.label == 0],data.y[data.label == 0],color ="pink")
plt.scatter(data.x[data.label == 1],data.y[data.label == 1],color = "purple")
plt.scatter(data.x[data.label == 2],data.y[data.label == 2],color = "blue")
plt.scatter(kmeans2.cluster_centers_[:,0],kmeans2.cluster_centers_[:,1],color = "yellow")
plt.show()
```
